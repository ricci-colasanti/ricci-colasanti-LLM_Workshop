```python
import requests
from tqdm import tqdm
import os

# Download model
url = "https://huggingface.co/bartowski/Meta-Llama-3-8B-Instruct-GGUF/resolve/main/Meta-Llama-3-8B-Instruct-Q4_K_M.gguf"
filename = "Meta-Llama-3-8B-Instruct-Q4_K_M.gguf"

print(f"Downloading {filename}...")
response = requests.get(url, stream=True)
total_size = int(response.headers.get('content-length', 0))

with open(filename, 'wb') as file, tqdm(
        desc=filename,
        total=total_size,
        unit='iB',
        unit_scale=True,
        unit_divisor=1024,
    ) as bar:
    for data in response.iter_content(chunk_size=1024):
        bar.update(len(data))
        file.write(data)

print(f"Download complete! File: {filename}")
```

```python
import time
from llama_cpp import Llama

# ------------------------------------------------------------------
# Load the model
# ------------------------------------------------------------------
print("Loading Mistral model…")
start = time.time()

llm = Llama(
    model_path="Meta-Llama-3-8B-Instruct-Q4_K_M.gguf",
    n_ctx=32768,          # 32 k token context
    n_threads=None,       # auto‑detect CPU cores
    verbose=False
)

print(f"Model loaded in {time.time() - start:.2f}s")

# ------------------------------------------------------------------
# Query the model
# ------------------------------------------------------------------
response = llm.create_chat_completion(
    messages=[{
        "role": "user",
        "content": "Explain what makes Mistral different from other LLMs in 2‑3 sentences."
    }],
    max_tokens=200,
    temperature=0.7
)

print(response["choices"][0]["message"]["content"])
```
